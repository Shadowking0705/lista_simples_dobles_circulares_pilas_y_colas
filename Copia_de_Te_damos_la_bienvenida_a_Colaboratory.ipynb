```
from re import A
class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.siguiente = None
        self.anterior = None

class Lista:
    def __init__(self):
        self.primero = None

    def inserta_primero_listaDoble(self, dato):
        nuevo = Nodo(dato)
        if self.primero is None:
          self.primero = nuevo
        else:
          nuevo.siguiente = self.primero
          self.primero.anterior = nuevo
          self.primero = nuevo

    def insertar_al_final(self, dato):
        nuevo_nodo = Nodo(dato)
        if self.primero == None:
            self.primero = nuevo_nodo
        else:
            actual = self.primero
            while actual.siguiente != None:
                actual = actual.siguiente
            actual.siguiente = nuevo_nodo
            nuevo_nodo.anterior = actual


    def insertar_despues_del_elemento(self, dato, datoB):
        nuevo = Nodo(dato)
        if self.primero is not None:
          aux = self.primero
          while aux.dato != datoB:
            aux = aux.siguiente
          nuevo.siguiente = aux.siguiente
          aux.siguiente.anterior = nuevo
          nuevo.anterior = aux
          aux.siguiente = nuevo
        else:
          self.primero = nuevo

    def insertar_antes_del_elemento(self, dato, datoB):
        nuevo = Nodo(dato)
        if self.primero is not None:
          aux = self.primero
          while aux.siguiente is not None and aux.siguiente.dato != datoB:
                aux = aux.siguiente
          if aux.siguiente is None:
                print(f'El dato {datoB} no se encuentra en la lista')
          nuevo.siguiente = aux.siguiente
          aux.siguiente.anterior = nuevo
          nuevo.anterior = aux
          aux.siguiente = nuevo
        else:
          return "no se puede agregar en una lista vacia"



    def eliminar_nodo_despues(self, datoB):
        if self.primero is None:
          return "No se puede eliminar un nodo en una lista vacía"

        aux = self.primero
        while aux is not None and aux.dato != datoB:
          aux = aux.siguiente

        if aux is None:
          return "Elemento no encontrado"

        if aux.siguiente is not None:
          nodo_a_eliminar = aux.siguiente
          aux.siguiente = aux.siguiente.siguiente
          aux.siguiente.anterior = aux
          nodo_a_eliminar.siguiente = None
          nodo_a_eliminar.anterior = None
        else:
          return "No hay nodo después del nodo con el dato proporcionado"


    def eliminar_nodo_antes(self, datoB):
        if self.primero is None:
            return "La lista está vacía"

        aux = self.primero
        while aux is not None and aux.dato != datoB:
            aux = aux.siguiente

        if aux is None or aux.anterior is None:
            return "El elemento no fue encontrado o es el primer nodo"

        if self.primero.siguiente.dato == datoB:
          nodo_a_eliminar = self.primero
          self.primero = self.primero.siguiente
          self.primero.anterior = None
          nodo_a_eliminar.siguiente = None
          return "Nodo antes eliminado"

        if aux.anterior is not None:
          nodo_eliminar = aux.anterior
          aux.anterior = aux.anterior.anterior
          aux.anterior.siguiente = aux
          nodo_eliminar.siguiente = None
          nodo_eliminar.anterior = None

        else:
          return "no hay un nodo antes del buscado"



    def eliminar_primero(self):
      if self.primero is None:
        return "lista vacia"
      if self.primero.siguiente is None:
        self.primero = None
      self.primero = self.primero.siguiente
      self.primero.anterior.siguiente = None
      self.primero.anterior = None


    def eliminar_por_el_final(self):
      if self.primero is None:
        return "lista vacia"
      elif self.primero.siguiente is None:
        self.primero = None
      else:
        aux = self.primero
        while aux.siguiente != None:
          aux = aux.siguiente
        aux.anterior.siguiente = None
        aux.anterior = None

    def imprimir_lista_doble(self):
        if self.primero == None:
            print("lista vacia")
        else:
            aux = self.primero
            while aux:
                print(aux.dato, end=" ")
                aux = aux.siguiente
            print()

mi_lista = Lista()
mi_lista.inserta_primero_listaDoble(45)
mi_lista.inserta_primero_listaDoble(5)
mi_lista.inserta_primero_listaDoble(10)
mi_lista.inserta_primero_listaDoble(80)
mi_lista.inserta_primero_listaDoble(100)
mi_lista.inserta_primero_listaDoble(90)
mi_lista.eliminar_por_el_final()
mi_lista.eliminar_por_el_final()
mi_lista.eliminar_por_el_final()
mi_lista.eliminar_por_el_final()
mi_lista.imprimir_lista_doble()
```

```
class Pila:
    def __init__(self):
        self.items = []

    def push(self, dato):
        self.items.append(dato)
        print(f"{dato} ha sido agregado a la pila")

    def esta_vacia(self):
        return len(self.items) == 0

# Ejemplo de uso
pila = Pila()
pila.push(5)
pila.push(10)
print("La pila está vacía:", pila.esta_vacia())
```

```
class Cola:
    def __init__(self):
        self.items = []

    def encolar(self, dato):
        self.items.insert(0, dato)
        print(f"{dato} ha sido encolado")

    def esta_vacia(self):
        return len(self.items) == 0

# Ejemplo de uso
cola = Cola()
cola.encolar(5)
cola.encolar(10)
print("La cola está vacía:", cola.esta_vacia())
```
